```python
import requests
from bs4 import BeautifulSoup

url = "https://soict.hust.edu.vn/can-bo/page/{number}"
links = []
for i in range(1, 10):

    response = requests.get(url.format(number=i))
    # print(response)
    if response.status_code != 200:
        break
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.text, "html.parser")

    # Find all div elements with the specific class
    divs = soup.find_all("div", class_="col small-12 large-3")

    # Extract href links from anchor tags within each div
    new_links = [a['href'] for div in divs for a in div.find_all("a", href=True)]
    # print(new_links)
    links.extend(new_links)
links = list(set(links))
```

```python
for link in links:
    print(link)
```
